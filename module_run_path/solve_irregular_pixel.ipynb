```python
import cv2


img = cv2.imread(r"D:\test3_error\20211023_t054904.jpg",0)
img_copy = img
img_t = img.T
dlist = []


# 橫向比較
shape = img.shape
ymax = shape[0]
xmax = shape[1]
#print("image shape:", ymax,xmax)

for i in range(ymax):    
    for j in range(xmax):
        
        n = int(img[i][j])
   
        if j+1 < xmax:
            r = int(img[i][j+1])
            d = abs(r-n)
            dlist.append(d)

# 縱向比較            
shape = img_t.shape
ymax = shape[0]
xmax = shape[1]
#print("image shape:", ymax,xmax)

for i in range(ymax):    
    for j in range(xmax):
        
        n = int(img_t[i][j])
  
        if j+1 < xmax:
            r = int(img_t[i][j+1])
            d = abs(r-n)
            dlist.append(d)
            

print("Total list length:", len(dlist)) 
print("Max differ pixel value:", max(dlist))
```

## 分析各點差異並返回四分位數

```python
import numpy as np
notice = np.percentile(dlist, 99.9)
print(notice)
```
